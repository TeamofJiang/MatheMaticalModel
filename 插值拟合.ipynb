### 插值

```python
import numpy as np
from scipy import interpolate as inp
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 3.14, 10)
y = np.sin(x)
```

```python
plt.plot(x, y)
```

```python
f1 = inp.interp1d(x, y, kind = "linear")
plt.plot(x, f1(x))
```

```python
f2 = inp.interp1d(x, y, kind = "cubic")
x2 = np.linspace(0, 3.14, 25)
plt.plot(x2, f2(x2),"*")
plt.plot(x, y , "+")
```

```python
f3 = inp.interp1d(x, y, kind = "cubic")
x3 = np.linspace(0, 3.14, 100)
plt.plot(x3, f3(x3))
```

**如上所示，插值的基本格式为  
   inp.interp1d(   
    &ensp; x, //插值节点的横坐标   
       &ensp; y, //插值节点的纵坐标   
       &ensp; kind = "cubic"  //插值方法   
)   
kind = "linear", "nearest", "cubic"**

**含有噪声点时的插值**

```python
x = np.linspace(0, 6.28, 50)
y = np.sin(x)
y += np.random.randn(50)

sp1 = inp.UnivariateSpline(x, y)
x2 = np.linspace(0, 6.28, 60)
plt.plot(x2, sp1(x2), "blue")

sp1.set_smoothing_factor(0)
plt.plot(x2, sp1(x2), "red")
```

### 拟合（最小二乘法）

```python
from scipy.optimize import leastsq
```

```python
x = np.linspace(0, 3.14*3, 50)
y = np.sin(x)
```

```python
def func(p, x):
    a0, a1, a2, a3, a4, a5 = p
    return a0 + a1*x +a2*x**2 + a3*x**3 + a4*x**4 + a5*x**5
```

```python
def error(p, x, y, s):
    print(s)
    return func(p, x) - y
```

```python
s = "Test"
p0 = [1, 1, 1, 1, 1, 1]
Para = leastsq(error, p0 ,args = (x, y, s))
```

```python
a0, a1, a2, a3, a4, a5 = Para[0]
plt.plot(x, a0 + a1*x + a2*x**2+ a3*x**3 + a4*x**4 + a5*x**5)
plt.plot(x,y)
```
